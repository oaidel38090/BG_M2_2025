
### Pourquoi notre site web ne fonctionne plus [http://IPdelaVM:88](http://192.168.56.110:88) ?

Créons à présent notre fichier myindex.html dans le répertoire local hôte /My/path/to/architectures_distribuees/docker/html :


Jetons un oeil sur notre site web [http://IPdelaVM:88](http://192.168.56.110:88)

Super notre site web fonctionne avec le contenu qu'on a indiqué dans notre répertoire local hôte.

Dans le module, nous utiliserons ce type d'attachement en particulier pour les fichiers de configuration des services utilisés.


Vous pouvez à présent supprimer l'ensemble des volumes, réseau et containers précédemment créés.

## 6 - Téléchargement d'une image

Par défaut, votre application Docker est configuré pour utiliser le catalogue d'image le Docker Registry connu sous le nom de docker hub ( https://hub.docker.com/ ).

Pour télécharger une image Docker  il vous suffit d'utiliser l'option pull de la commande docker.

A titre d'information, ce qui suit n'est pas obligatoire pour la suite des TP mais cela vous permet de tester le téléchargement d'un container à partir du registry docker.

La commande ci-dessous téléchargera l'image hello-world:latest


### Vérifier que  l'image hello-world est à présent bien disponible sur votre VM :

### Instancier un container à partir de l'image hello-world:latest sans le mode detach ?

### Pourquoi votre container n'est pas en cours d'exécution ?

Pour supprimer une image, on utilisera la commande `docker image rm hello-world:latest`

--- 

Notez que la suppression d'une image n'est possible que si celle-ci n'est pas utilisée par un container quelque soit son statut (stopped / Running ).

---

### Procéder à la suppression de l'image httpd:latest en expliquant les différentes étapes ?

# Compléter votre réponse

## 7 - Docker compose



Docker compose est un orchestrateur multi-containers conçu pour faciliter le déploiement d'un parc de container.

Toutes les options que nous avons vu avec la commande docker peuvent être centralisée dans un fichier docker-compose.yml  de manière à démarrer automatiquement nos containers avec la configuration souhaitée.

Le fichier docker-compose dispose de plusieurs sections dont 3 nous intéressent particulièrement :
    - services contient la définition des instantiation de nos containers
    - networks la définition des réseaux 
    - volumes la définition des volumes


Pour démarrer un simple service nginx il suffit de créer un fichier de configuration docker-compose-test.yml

```
version: "3"

services:

  jupyter: # on instancie un service dont le nom est web
    image: guizmo38/ynov_training_ad:v1.0 # à partir de cette image
    container_name: 'jupyter' # et on associe le nom jupyter à notre container
    hostname : jupyter # Le nom affiché dans le container sera jupyter au lieu d'une valeur aléatoire 06587b37c19d
    ports:
      - "8888:8888"  # On définit la liste des ports publiés pour ce service
    volumes:
      - ./jupyter:/home/jovyan/training # et on attache le répertoire hôte ./jupyter au répertoire  /home/jovyan/training du container. Notez que l'on peut indiquer un chemin relatif avec docker compose.
    networks:
      - bigdata   # le réseau auquel est rattaché le service

# La section network créera automatiquement les réseaux indiqués
networks:
  bigdata:


# La section volume créera automatiquement les volumes indiqués
volumes:
  hadoop_namenode:
```

### A partir de l'exemple ci-dessus créer un fichier docker-compose-test.yml qui permettra de démarrer notre fameux service web avec la configuration suivante :
* Publication sur le port 90
* Un volume mywebdisk attaché au point de montage  le répertoire /usr/share/nginx/html
* un réseau webnet
* un container dont le nom sera myweb
* on pourra préciser comme nom de host pour le container myweb01
</font>

Pour tester votre fichier docker-compose-test.yml, il vous suffit de lancer la commande suivante :


Pour arrêter un ou plusieurs containers :

```
# pour arrêter un service
docker compose -f docker-compose-test.yml stop web
# pour arrêter tous les services spécifiés dans le fichier de configuration
docker compose -f docker-compose-test.yml stop
```

Pour démarrer un ou plusieurs containers :

```
# pour démarrer un service
docker compose -f docker-compose-test.yml start web
# pour arrdémarrerêter tous les services spécifiés dans le fichier de configuration
docker compose -f docker-compose-test.yml start
```

Vous l'aurez compris on retrouvera toutes les options proposées sur docker.
Bien entendu, l'utilisation des commandes docker est toujours possible mais à éviter.

### Essayez à présent d'ajouter dans votre fichier de configuration docker-compose-test.yml un second container web2 avec les indications suivantes :

* Publication sur le port 92
* Un volume mywebdisk attaché au point de montage  le répertoire /usr/share/nginx/html
* un réseau webnet
* un container dont le nom sera myweb2
* on pourra préciser comme nom de host pour le container myweb02

Puis lancez l'instanciation des 2 containers en une seule commande .

</font>
  

Instanciation des containers


docker compose -f docker-compose-test.yml up -d


###  Comment pouvez vous supprimer tous les containers et les réseaux contenus dans votre fichier de configuration ?

--- 

Regarder par ici : https://docs.docker.com/engine/reference/commandline/compose_down/

---



### Que réalise le docker-compose suivant sans exécuter aucune commande ? 

Bravo, vous avez terminé ce premier TP qui je l'espère vous a permis de mieux comprendre Docker et comment l'utiliser pour vos propres besoins.


Pour supprimer tous les containers liés à ce TP, nous pourrons utiliser les commandes suivantes :
