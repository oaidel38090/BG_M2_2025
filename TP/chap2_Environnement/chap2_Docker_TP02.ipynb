
### Pourquoi notre site web ne fonctionne plus [http://IPdelaVM:88](http://192.168.56.110:88) ?

Crons  prsent notre fichier myindex.html dans le rpertoire local hte /My/path/to/architectures_distribuees/docker/html :


Jetons un oeil sur notre site web [http://IPdelaVM:88](http://192.168.56.110:88)

Super notre site web fonctionne avec le contenu qu'on a indiqu dans notre rpertoire local hte.

Dans le module, nous utiliserons ce type d'attachement en particulier pour les fichiers de configuration des services utiliss.


Vous pouvez  prsent supprimer l'ensemble des volumes, rseau et containers prcdemment crs.

## 6 - Tlchargement d'une image

Par dfaut, votre application Docker est configur pour utiliser le catalogue d'image le Docker Registry connu sous le nom de docker hub ( https://hub.docker.com/ ).

Pour tlcharger une image Docker  il vous suffit d'utiliser l'option pull de la commande docker.

A titre d'information, ce qui suit n'est pas obligatoire pour la suite des TP mais cela vous permet de tester le tlchargement d'un container  partir du registry docker.

La commande ci-dessous tlchargera l'image hello-world:latest


### Vrifier que  l'image hello-world est  prsent bien disponible sur votre VM :

### Instancier un container  partir de l'image hello-world:latest sans le mode detach ?

### Pourquoi votre container n'est pas en cours d'excution ?

Pour supprimer une image, on utilisera la commande `docker image rm hello-world:latest`

--- 

Notez que la suppression d'une image n'est possible que si celle-ci n'est pas utilise par un container quelque soit son statut (stopped / Running ).

---

### Procder  la suppression de l'image httpd:latest en expliquant les diffrentes tapes ?

# Complter votre rponse

## 7 - Docker compose



Docker compose est un orchestrateur multi-containers conu pour faciliter le dploiement d'un parc de container.

Toutes les options que nous avons vu avec la commande docker peuvent tre centralise dans un fichier docker-compose.yml  de manire  dmarrer automatiquement nos containers avec la configuration souhaite.

Le fichier docker-compose dispose de plusieurs sections dont 3 nous intressent particulirement :
    - services contient la dfinition des instantiation de nos containers
    - networks la dfinition des rseaux 
    - volumes la dfinition des volumes


Pour dmarrer un simple service nginx il suffit de crer un fichier de configuration docker-compose-test.yml

```
version: "3"

services:

  jupyter: # on instancie un service dont le nom est web
    image: guizmo38/ynov_training_ad:v1.0 #  partir de cette image
    container_name: 'jupyter' # et on associe le nom jupyter  notre container
    hostname : jupyter # Le nom affich dans le container sera jupyter au lieu d'une valeur alatoire 06587b37c19d
    ports:
      - "8888:8888"  # On dfinit la liste des ports publis pour ce service
    volumes:
      - ./jupyter:/home/jovyan/training # et on attache le rpertoire hte ./jupyter au rpertoire  /home/jovyan/training du container. Notez que l'on peut indiquer un chemin relatif avec docker compose.
    networks:
      - bigdata   # le rseau auquel est rattach le service

# La section network crera automatiquement les rseaux indiqus
networks:
  bigdata:


# La section volume crera automatiquement les volumes indiqus
volumes:
  hadoop_namenode:
```

### A partir de l'exemple ci-dessus crer un fichier docker-compose-test.yml qui permettra de dmarrer notre fameux service web avec la configuration suivante :
* Publication sur le port 90
* Un volume mywebdisk attach au point de montage  le rpertoire /usr/share/nginx/html
* un rseau webnet
* un container dont le nom sera myweb
* on pourra prciser comme nom de host pour le container myweb01
</font>

Pour tester votre fichier docker-compose-test.yml, il vous suffit de lancer la commande suivante :


Pour arrter un ou plusieurs containers :

```
# pour arrter un service
docker compose -f docker-compose-test.yml stop web
# pour arrter tous les services spcifis dans le fichier de configuration
docker compose -f docker-compose-test.yml stop
```

Pour dmarrer un ou plusieurs containers :

```
# pour dmarrer un service
docker compose -f docker-compose-test.yml start web
# pour arrdmarrerter tous les services spcifis dans le fichier de configuration
docker compose -f docker-compose-test.yml start
```

Vous l'aurez compris on retrouvera toutes les options proposes sur docker.
Bien entendu, l'utilisation des commandes docker est toujours possible mais  viter.

### Essayez  prsent d'ajouter dans votre fichier de configuration docker-compose-test.yml un second container web2 avec les indications suivantes :

* Publication sur le port 92
* Un volume mywebdisk attach au point de montage  le rpertoire /usr/share/nginx/html
* un rseau webnet
* un container dont le nom sera myweb2
* on pourra prciser comme nom de host pour le container myweb02

Puis lancez l'instanciation des 2 containers en une seule commande .

</font>
  

Instanciation des containers


docker compose -f docker-compose-test.yml up -d


###  Comment pouvez vous supprimer tous les containers et les rseaux contenus dans votre fichier de configuration ?

--- 

Regarder par ici : https://docs.docker.com/engine/reference/commandline/compose_down/

---



### Que ralise le docker-compose suivant sans excuter aucune commande ? 

Bravo, vous avez termin ce premier TP qui je l'espre vous a permis de mieux comprendre Docker et comment l'utiliser pour vos propres besoins.


Pour supprimer tous les containers lis  ce TP, nous pourrons utiliser les commandes suivantes :
